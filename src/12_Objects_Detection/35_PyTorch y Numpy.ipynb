## **Creando una instancia de `torch.Tensor` a partir de datos**

```
datos = np.array([1,2,3,4,5], dtype=np.int32)
print(datos)
print(datos.dtype)
```

```
t = torch.Tensor(datos) #constructor
print(t)
print(t.dtype)
```

```
t = torch.tensor(datos) #factory
print(t)
print(t.dtype)
```

```
t = torch.as_tensor(datos) #factory
print(t)
print(t.dtype)
```

```
t = torch.from_numpy(datos) #factory
print(t)
print(t.dtype)
```

```
torch.get_default_dtype()
```
